```python
# Here import the libraries which will be used
import matplotlib.pyplot as plt # Plotting library
import numpy as np # Numerical object library
```

```python
# Define scalar variables
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0
print(x)
print(y)
print(z)
print(type(x))
print(type(y))
print(type(z))
```

```python
# Define boolean variables
A = True
B = False

print(A and B)
print(A and (not B))
```

```python
# Define a string object
name = 'Camille'
print(name)
```

```python
# Define a list
numbers_and_letters = [1,2,3,4,5,'a','b','c']
print(numbers_and_letters)
```

```python
# Access values in a list
print(numbers_and_letters[0]) # ! Python indices start from 0

print(numbers_and_letters[2:4]) # ! Print values from 3rd to 5th

print(numbers_and_letters[-2:]) # ! Print two last values
```

```python
# Define a vector
x = np.arange(0,1,0.1)
print(x)
```

```python
# Define a for loop
for i in range(10):
    print(i)    # print index
    print(x[i]) # access elements in x
```

```python
# Define and evaluate a function
x = np.arange(-2,6,0.1)
def func(x):
    return 6*x - 3*x**2

f_x = func(x)
print(f_x)
```

```python
# Plot a function
fig = plt.figure(dpi=300)
plt.plot(x,func(x))
plt.grid()
plt.scatter(np.array([1]),func(np.array([1])),color='r')
plt.xlabel('x',fontsize=20)
plt.ylabel('f(x)',fontsize=20)
```

```python
# Define a function with free parameters

x = np.arange(-2,6,0.1)
def func(x,a=6):
    return a*x - 3*x**2

f_x = func(x,a=2)
# Plot a function
fig = plt.figure(dpi=300)
plt.plot(x,func(x,6),label='a = 6')
plt.plot(x,func(x,2),label='a = 2')
plt.grid()
plt.xlabel('x',fontsize=20)
plt.ylabel('f(x)',fontsize=20)
plt.legend()
```
