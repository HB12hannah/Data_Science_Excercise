```python
#wollte ausprobieren ob ich das auch an anfang schreiben kann statt in die klammmer bei Func
A = 2 #Amplitude, in d) mit 2 plotten  
T = 6 #Periode
t_0 = 3 #Phasenverschiebung, 3 Monate die Funktion zurÃ¼ck schieben in e) (t_0=3)
d = 3 #value 

steps = [0.25, 1, 3.47, 6.13]
#wenn ich 1 nehme hab ich einen Wert fÃ¼r den Monat, wenn 0.1 dann 10 Werte im Monat und die Kurve verlÃ¤uft glatter, wissenschaftlich glaube immer bessert mehrer Punkte zu haben

#Schleife mit Chat aber selbst angeordnet im Skript, hÃ¤tte auch bei jedem Plot den step Ã¤ndern kÃ¶nnen und abspeichern kÃ¶nnen 

for step in steps:
    t = np.arange(0, 36, step)    
    def func(t):
        return A *np.cos(2*np.pi/T * (t - t_0)) + d

    fig = plt.figure(figsize=(10, 6))
    plt.plot(t, func(t))
    plt.grid()
    plt.xlabel('months (t)',fontsize=20)
    plt.ylabel('g(t)',fontsize=20)
    plt.legend([f'step size = {step}'])

    mean_value = np.mean(func(t))
    print(f'Mean value for step size {step}: {mean_value}')
```
